```python
import pandas as pd
stock_data = pd.read_csv('Data/stock data/sh600000.csv', parse_dates=[1])
stock_data[:5]
```

```python
"""
B         business day frequency
C         custom business day frequency (experimental)
D         calendar day frequency
W         weekly frequency
M         month end frequency
SM        semi-month end frequency (15th and end of month)
BM        business month end frequency
CBM       custom business month end frequency
MS        month start frequency
SMS       semi-month start frequency (1st and 15th)
BMS       business month start frequency
CBMS      custom business month start frequency
Q         quarter end frequency
BQ        business quarter endfrequency
QS        quarter start frequency
BQS       business quarter start frequency
A         year end frequency
BA, BY    business year end frequency
AS, YS    year start frequency
BAS, BYS  business year start frequency
BH        business hour frequency
H         hourly frequency
T, min    minutely frequency
S         secondly frequency
L, ms     milliseconds
U, us     microseconds
N         nanoseconds
"""
period_type = 'W'
```

```python
# 将date设定为index
date_stock_data = stock_data.set_index('date')
date_stock_data[:5]
```

```python
# 进行转换，周线的每个变量都等于那一周中最后一个交易日的变量值
period_stock_data = date_stock_data.resample(period_type).last()
period_stock_data[:5]
```

```python
# 周线的open等于那一周中第一个交易日的open
period_stock_data['open'] = date_stock_data['open'].resample(period_type).first()
period_stock_data[['open']][:5]
```

```python
# 周线的high等于那一周中high的最大值
period_stock_data['high'] = date_stock_data['high'].resample(period_type).max()
period_stock_data[['high']][:5]
```

```python
# 周线的low等于那一周中low的最小值
period_stock_data['low'] = date_stock_data['low'].resample(period_type).min()
period_stock_data[['low']][:5]
```

```python
# 周线的volume和money等于那一周中volume和money各自的和
period_stock_data['volume'] = date_stock_data['volume'].resample(period_type).sum()
period_stock_data['money'] = date_stock_data['money'].resample(period_type).sum()
period_stock_data[['money', 'volume']][:5]
```

```python
# 计算周线turnover
period_stock_data['turnover'] = period_stock_data['volume'] / (period_stock_data['traded_market_value']/period_stock_data['close'])
period_stock_data[['turnover']][:5]
```

```python
# 股票在有些周一天都没有交易，将这些周去除
period_stock_data = period_stock_data[period_stock_data['code'].notnull()]
period_stock_data.reset_index(inplace=True)
period_stock_data[:5]
```
